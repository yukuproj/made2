предложите метод обучения перестановки символов в этом задании, основанный на MCMC-сэмплировании, но по-прежнему работающий на основе статистики биграмм;

Идея алгоритма - определяем случайную смену способа кодирования. Далее считаем скор расшифрованного сообщения, используя статистику биграмм из корпуса. Скор при этом считается как так - проходим циклом по всем биграммам сообщения и для каждой биграммы считаем логарифм частота этой биграммы из корпуса, полученные логарифмы суммируем. Получается своего рода функция правдоподобия. Полученный скор случайного изменения способа кодирования далее сравниваем со скором сообщения первоначального кодирования. Если новый скор будем больше скора до изменения кодирования, то применяем этот новый способ кодирования, если же новый скор меньше предыдущего, то подбрасываем монетку с вероятностью = экспонента от разности скоров, т.е. с вероятностью равной отношению вероятностей получить сообщение после применения нового кодирования к вероятности получить сообщения согласно первоначальному способу кодирования. Фактически это то самое отношение вероятностей в методе MCMC. Если монетка выпала удачей, то принимаем смену способа кодирования, в противном случае, изменение способа кодирования игнорируем. Далее повторяем процесс изменения кодирования. Делаем несколько тысяч итераций и в какой-то момент можно остановиться поскольку сообщение уже раскодируется.

```python
alphabet_dic = {}
for i in range(len(alphabet)):
    alphabet_dic[alphabet[i]] = i
```

```python
M = np.zeros((len(alphabet), len(alphabet)))
```

```python
for key in bigrams.keys():
    M[alphabet_dic[key[0]]][alphabet_dic[key[1]]] = bigrams[key]
```

```python
encoding3 = {}
for key in decoding.keys():
    encoding3[decoding[key]] = key
```

```python
decoding_alphabet = np.array([encoding3[alphabet[i]] for i in range(len(alphabet))])
```

```python
decoding3 = decoding.copy()
```

```python
def score(text, decoding, M):
    score = 0.
    for i in range(len(text) - 2):
        w1 = text[i:i+2][0]
        w2 = text[i:i+2][1]
        score += np.log( M[alphabet_dic[decoding[w1]]][alphabet_dic[decoding[w2]]])
    return score
```

```python
def change_decoder(alphabet, decoding_alphabet):
    idx_1, idx_2 =  np.random.choice(range(len(alphabet)), size=2, replace=False)   
    decoding_alphabet[idx_1], decoding_alphabet[idx_2] = decoding_alphabet[idx_2], decoding_alphabet[idx_1]
    decoding = {}
    for i in range(len(alphabet)):
        decoding[decoding_alphabet[i]] = alphabet[i]
    return decoding, decoding_alphabet
```

```python
for epoch in range(20000):
    score1 = score(test_coding, decoding3, M)
    decoding3_next, decoding_alphabet_next = change_decoder(alphabet.copy(), decoding_alphabet.copy())
    score2 = score(test_coding, decoding3_next, M)
    rez = False
    if score1 < score2:
        rez = True        
    else:
        p = np.exp(score2 - score1)
        x = np.random.binomial(1, p, 1)
        if x == 1:
          rez = True
    if rez:
          decoding3 = decoding3_next.copy()
          decoding_alphabet = decoding_alphabet_next.copy()
    if epoch % 1000 == 0:
        score1 = score(test_coding, decoding3, M)
        test_decoding = ''
        for s in test_coding:
            test_decoding += decoding3[s]
        print('Current score = ' + str(score1))
        print(test_decoding)
```
