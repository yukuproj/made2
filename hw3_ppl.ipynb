```python
alphabet_dic = {}
for i in range(len(alphabet)):
    alphabet_dic[alphabet[i]] = i
```

```python
M = np.zeros((len(alphabet), len(alphabet)))
```

```python
for key in bigrams.keys():
    M[alphabet_dic[key[0]]][alphabet_dic[key[1]]] = bigrams[key]
```

```python
decoding_alphabet = np.array([encoding3[alphabet[i]] for i in range(len(alphabet))])
```

```python
decoding3 = decoding.copy()
```

```python
def score(text, decoding, M):
    score = 0.
    for i in range(len(text) - 2):
        w1 = text[i:i+2][0]
        w2 = text[i:i+2][1]
        score += np.log( M[alphabet_dic[decoding[w1]]][alphabet_dic[decoding[w2]]])
    return score
```

```python
def change_decoder(alphabet, decoding_alphabet):
    idx_1 = np.random.randint(0, len(alphabet))
    idx_2 = np.random.randint(0, len(alphabet))
    while idx_1 == idx_2:
        idx_2 = np.random.randint(0, len(alphabet))
    decoding_alphabet[idx_1], decoding_alphabet[idx_2] = decoding_alphabet[idx_2], decoding_alphabet[idx_1]
    decoding = {}
    for i in range(len(alphabet)):
        decoding[decoding_alphabet[i]] = alphabet[i]
    return decoding, decoding_alphabet
```
